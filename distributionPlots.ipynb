```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.load_dataset('tips')
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.histplot(df['total_bill']) #histogram
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],kde=True)
```

```python
#jointplot (2 data ko ek saath plot krna ho ek graph m)
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter') #scatter plot bnega
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hist')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.pairplot(df,hue='sex',palette='rainbow') #pairplot(yeh graph show krdeta h numeric value jisme hoti h) hue(yeh categories vise distribute krdeta h) palette(color scheme of a graph)
```

```python
sns.pairplot(df,hue='size',palette='rainbow') 
```
